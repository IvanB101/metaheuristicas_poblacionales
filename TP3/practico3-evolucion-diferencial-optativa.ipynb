El código corespondiente a la solución se encuentra en el archivo adjunto `de_2.py`.

En base a las ejecuciones realizadas se pudo observar que el valor de $F$ es inversamente proporcional a la velocidad con la que se encuentra la solución óptima. Esto se debe a que el valor de un mayor valor de $F$ da mayor peso a la diferencia de vectores, probablemente mejorando la exploración y empeorando la explotación. Dado que el decremente de cualquier $x_i$ resulta en el decremento de la función objetivo, la mayor explotación encuentra más rápido la solución, teniendo en cuenta que no existen otros mínimos locales en los que se pueda estabilizar.

Por otra parte un valor de $Cr$ para disminuir la velocidad con la que se encuentra la solución. Esto puede deberse a la menor variabilidad entre generaciones como consecuencia del menor valor de $Cr$. Dicho esto, la variación en el número de generaciones requerido para encontrar la solución, o acercarce a alguna distancia concreta de la misma, varía mucho más levemente con cambios de $Cr$ que con cambios de $F$.

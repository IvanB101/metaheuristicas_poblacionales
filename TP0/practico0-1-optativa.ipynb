$\mathbf{Ejercicio \,\,piloto}$: Crear un arreglo $\mathbf{A}$ de $100$ componentes enteras con valores generados de manera aleatoria en el rango $[-10,10]$. Imprimir la cantidad de componentes de $\mathbf{A}$ con valores positivos y la suma acumulada de las mismas.

$\mathbf{Ejercicio} \,\,1$: Generar un arreglo $\mathbf{A}$ de 100 componentes con valores aleatorios en el rango continuo $[1,100]$ e imprimir $\mathbf{A}$. Luego, para aquellas componentes de $\mathbf{A}$ con valores entre $20$ y $60$, darles el valor $0$. Imprimir $\mathbf{A}$ nuevamente.

```python
from random import uniform

seed(10)

A = [uniform(1,100) for _ in range(100)]
print("Arreglo inicial:")
print(A)

A = [elem if elem <=20 or elem >= 60 else 0 for elem in A]
print("Arreglo modificado:")
print(A)
```

$\mathbf{Ejercicio} \,\,2$: Idem ejercicio 1, pero generar un arreglo $\mathbf{B}$ con las componentes que cumplan la condición y eliminarlas del arreglo $\mathbf{A}$, es decir que  $\mathbf{A}$ debe quedar sólamente con todas las componentes que no fueron eliminadas. Esto implicará un cambio en el tamaño del arreglo $\mathbf{A}$. Imprimir los arreglos  $\mathbf{A}$ y  $\mathbf{B}$

```python
A = [uniform(1,100) for _ in range(100)]
print("Arreglo inicial:")
print(A)

def check(n: float):
    return n <=20 or n >= 60

B = [a for a in A if check(a)]
print("B:")
print(B)
A = [a for a in A if not check(a)]
print("A:")
print(A)
```

```python
matrix = list[list[int]]

def rand_matrix(rows: int, columns: int) -> matrix:
    return [[randint(1,5) for _ in range(columns)] for _ in range(rows)]

def mult_matrix(A: matrix, B: matrix) -> matrix:
    return [[sum([elem * row_B[col] for elem, col in zip(row_A, range(len(row_A)))]) for row_B in B] for row_A in A]

# Solucion alternativa sin listas por compresion
# def alt_mult_matrix(A: matrix, B: matrix) -> matrix:
#     res = []
#     for row_A in A:
#         row = []
#         for row_B in B:
#             sum = 0
#             for elem, col in zip(row_A, range(len(row_A))):
#                 sum += elem * row_B[col]
#             row.append(sum)
#         res.append(row)
#     return res

A = rand_matrix(20, 10)
B = rand_matrix(10, 30)

print(mult_matrix(A, B))
```

$\mathbf{Ejercicio} \,\,6$: Idem ejercicio anterior pero generando una matriz
$\mathbf{C}=\mathbf{A} \times \mathbf{B}$. Luego, eliminar de la matriz $\mathbf{C}$
las columnas con índice par y generar un nueva matriz $\mathbf{D}$ con las columnas
eliminadas de $\mathbf{C}$. Imprimir $\mathbf{C}$ y $\mathbf{D}$.

```python
A = rand_matrix(20, 10)
B = rand_matrix(10, 30)

C = mult_matrix(A, B)
D = []

for row in C:
    row_D = []
    # Se comienza removiendo el primer elemento, conderando que 0 es par.
    for index in range(len(row) // 2):
        row_D.append(row.pop(index))
    D.append(row_D)

print("C: ", C)
print("D: ", D)
```

$\mathbf{Ejercicio} \,\,7$: Dados un arreglo $\mathbf{T}$ de $100$ componentes con valores
generados aleatoriamente en el rango continuo $[-20,20]$ y un arreglo $\mathbf{Mask}$
de $100$ componentes de tipo booleano con valores generados aleatoriamente. Se pide,
dividir el arreglo $\mathbf{T}$ en dos arreglos $\mathbf{P}$ y $\mathbf{N}$ según los
valores dados por $\mathbf{Mask}$. Imprimir todos los arreglos al final.

```python
T = [uniform(-20, 20) for _ in range(100)]
Mask = [random() < 0.5 for _ in range(100)]

P = []
N = []

for elem, cond in zip(T, Mask):
    if(cond):
        P.append(elem)
    else:
        N.append(elem)

print("T: ", T)
print("Mask: ", Mask)
print("P: ", P)
print("N: ", N)
```

$\mathbf{Ejercicio} \,\,8$: Dados 500 puntos en la plano (tipo flotante) generados
aleatoriamente en el conjunto $\{(x,y) / x,y \in [-100,100]\}$, determinar e informar
el número de puntos encontrados en el 1er, 2do, 3er y 4to cuadrante respectivamente.

```python
def get_cuadrant_2D(p: list[float]):
    x, y = p
    return (y < 0) << 1 | (x > 0) ^ (y > 0)

points = [[uniform(-100, 100), uniform(-100, 100)] for _ in range(500)]

count = [0 for _ in range(4)]

for p in points:
    count[get_cuadrant_2D(p)] += 1


print("Numero de puntos:")
for i in range(4):
    print('     Cuadrante {}: {}'.format(i + 1, count[i]))
```

$\mathbf{Ejercicio} \,\,9$: Idem ejercicio anterior, reportando tambien el punto medio
de cada uno los cuatro cuadrantes.

```python
points = [[uniform(-100, 100), uniform(-100, 100)] for _ in range(500)]

count = [0 for _ in range(4)]
acc = [[0., 0.] for _ in range(4)]

for p in points:
    cuad = get_cuadrant_2D(p)
    count[cuad] += 1
    acc[cuad][0] += p[0]
    acc[cuad][1] += p[1]

print("Numero de puntos:")
for i in range(4):
    print('     Cuadrante {}: {}'.format(i + 1, count[i]))

print("Punto medio:")
for i in range(4):
    print('     Cuadrante {}: {}'.format(i + 1, [acc / len(points) for acc in acc[i]]))
```

$\mathbf{Ejercicio} \,\,10$: Idem ejercicio anterior, pero generando en forma aleatoria
puntos en el conjunto $ \{(x,y,z) / x,y,z \in [-100,100]\}$ y reportando los puntos medios
de cada uno de los octantes.

```python
def get_cuadrant_3D(p: list[float]):
    x, y, z = p
    return (z < 0) << 2 | (y < 0) << 1 | (x > 0) ^ (y > 0)

points = [[uniform(-100, 100), uniform(-100, 100), uniform(-100, 100)] for _ in range(500)]

count = [0 for _ in range(8)]
acc = [[0., 0., 0.] for _ in range(8)]

for p in points:
    cuad = get_cuadrant_3D(p)
    count[cuad] += 1
    acc[cuad][0] += p[0]
    acc[cuad][1] += p[1]
    acc[cuad][2] += p[2]

print("Numero de puntos:")
for i in range(8):
    print('     Cuadrante {}: {}'.format(i + 1, count[i]))

print("Punto medio:")
for i in range(8):
    print('     Cuadrante {}: {}'.format(i + 1, [acc / len(points) for acc in acc[i]]))
```

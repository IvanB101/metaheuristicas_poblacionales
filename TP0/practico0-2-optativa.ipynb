$\mathbf{Ejercicio} \,\,1$: Una de la facilidades que permite IPython está relacionada a la escritura de los comandos en linea. Esto es interesante dado que al ser interactivo, mientras más fácil sea escribir los comandos, más eficiente será la interacción del ususario con el intérprete Python. Indique cuáles son dichas facilidades y describa brevemenente cada de una de ellas.

#### Respuesta
Una facilidad qur brinda IPython para la escritura de comandos de línea son los atajos de teclado, dentro de los cuales tenemos para:
- Navegación: para mover el cursor a distintasposiciones.
- Manipulación de texto: como copiar, cortar ypegar.

Otra características de suma utilidad es el autocompletado. Presionando `tab` IPython intenta completar lo que está escribiendo actualmente, haciendo uso de la información disponible.

$\mathbf{Ejercicio} \,\,3$: La funciones mágicas proveen una alternativa para incrementar la productividad por parte del usuario. Explique qué son las funciones mágicas y para qué han sido propuestas en el entorno de IPython.

#### Respuesta
Las funciones mágicas son funciones provistas por IPython, las cuales están antecididas por `%`. Han sido propuestas en el entorno de python problemas comunmente econtrados en el análisis de datos, tales como:
- Pegar bloques de código
- Ejecutar código externo
- Cronometrar el tiempo de ejecución

$\mathbf{Ejercicio} \,\,4$: ¿Cuáles son los mecanismos que provee IPython para acceder al historial de Entrada/Salida?

#### Respuesta
Para acceder al historial de E/S, IPython hace uso de dos objetos:
- `In`, una lista con las entradas
- `Out`, un diccionario con la salida, si es que hubo alguna, correspondiente al número de cada entrada.
Además de acceder los elementos de estas colecciones mediante índices, podemos acceder al último output con `_`, al penúltimo con `__` y al antepenúltimo con `___`.

$\mathbf{Ejercicio} \,\,5$: Idem ejercicio anterior para la facilidad de Exploración de ojetos (e.g., documentación, archivos de código, variables de entorno, etc.

### Respuesta
Para la exploración de objetos podemos colocar `?` despues del nombre del objeto en la terminal, esto nos muestra el `docstring` del objeto. También podemos colocar `??`, lo que nos brinda más información, como el código fuente del objeto, si está disponible.
Adicionalmente existen algunos comandos mágicos que podemos utilizar:
- `pdef`: muestra la definición de una función.
- `pdoc`: muestra el `docstring` de un objeto.
- `psource`: muestra el código fuente de un objeto.
- `pfile`: muestra el código fuente del archivo donde el objeto es definido.

$\mathbf{Ejercicio} \,\,6$: Tener la posibilidad de acceder a información respecto a las Sesiones realizadas en IPython puede ser de gran ayuda. Describa brevemente la manera en que esta facilidad es ofrecida por IPython.

#### Respuesta
Para obtener información sobre las sesiones de IPython, este brinda funciones mágicas, como:

- `%history`: nos permite ver el historial de comandos.
- `%dhist`: nos permite ver el historial de directorios visitados.
- `%whos`: brinda información sobre las variables en la sesión actual
- `%env`: muestra información sobre las variables de entorno.

Además también puede ser de utilidad el comando:
- `%logstart`: nos permite iniciar una sesión a partir de la información de una sesión previa, almacenada en un log.

$\mathbf{Ejercicio} \,\,7$: Finalmente (aunque hay otras facilidades adicionales brindadas por IPython que pueden ser consultadas en la bibliografía) la posibilidad de Editar y Depurar código es de gran importancia para realizar un desarrollo. Analizar cuáles son dichas facilidades y en particular, analizar aquella brindada para medir los tiempos de ejecución (facilidad importante para realizar profiling de los programas y/o trozos de código del mismo.

#### Respuesta
Facilidades para depurar tenemos:
- Control de excepciones: podemos controlar la cantidad de información brindada por el intérprete cuando ocurre una excepción.
- Depuración interactiva: existen varios comandos a nuestra disposión para poder depurar de forma interactiva.
- Medición de tiempo de ejecución

##### Medición de tiempo de ejecución
Para esto IPython cuenta con los siguiente comandos mágicos:
- `%time`: tiempo de ejecución de una única sentencia
- `%timeit`: tiempor repetido de ejeción de una única sentencia, para tener mayor precisión.
- `%prun`: ejecutar el código con el perfilador.
- `%lprun`: ejecutar el código con el perfilador línea por línea.
- `%memit`: mide el uso de memoria de una única sentencia.
- `%mprun`: ejecutar el código con el perfilador de memoria línea por línea.
